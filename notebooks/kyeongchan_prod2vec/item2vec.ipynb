```python
# ì˜í™” ì •ë³´ ë¡œë”©(10197 ì‘í’ˆ)
# movie_idì™€ ì œëª©ë§Œ ì‚¬ìš©
m_cols = ["movie_id", "title", "genre"]
movies = pd.read_csv(
    os.path.join(data_loader.data_path, "movies.dat"), names=m_cols, sep="::", encoding="latin-1", engine="python"
)
# genreë¥¼ list í˜•ì‹ìœ¼ë¡œ ì €ì¥í•œë‹¤
movies["genre"] = movies.genre.apply(lambda x: x.split("|"))

# ì‚¬ìš©ìê°€ ë¶€ì—¬í•œ ì˜í™”ì˜ íƒœê·¸ ì •ë³´ë¥¼ ë¡œë”©í•œë‹¤
t_cols = ["user_id", "movie_id", "tag", "timestamp"]
user_tagged_movies = pd.read_csv(
    os.path.join(data_loader.data_path, "tags.dat"), names=t_cols, sep="::", engine="python"
)
# tagë¥¼ ì†Œë¬¸ìë¡œ í•œë‹¤
user_tagged_movies["tag"] = user_tagged_movies["tag"].str.lower()
movie_tags = user_tagged_movies.groupby("movie_id").agg({"tag": list})

# íƒœê·¸ ì •ë³´ë¥¼ ê²°í•©í•œë‹¤
movies = movies.merge(movie_tags, on="movie_id", how="left")

# í‰ê°€ ë°ì´í„°ë¥¼ ë¡œë”©í•œë‹¤
r_cols = ["user_id", "movie_id", "rating", "timestamp"]
ratings = pd.read_csv(os.path.join(data_loader.data_path, "ratings.dat"), names=r_cols, sep="::", engine="python")

# user ìˆ˜ë¥¼ num_usersë¡œ ì¤„ì¸ë‹¤
valid_user_ids = sorted(ratings.user_id.unique())[: data_loader.num_users]
ratings = ratings[ratings.user_id <= max(valid_user_ids)]

# ìœ„ ë°ì´í„°ë¥¼ ê²°í•©í•œë‹¤
movielens_ratings = ratings.merge(movies, on="movie_id")
```
