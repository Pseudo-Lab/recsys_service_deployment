```python
def check_director_exist(cast):
    if not cast:
        return 0
    cast = eval((cast))
    director_cnt = 0
    for person_dict in cast:
        if person_dict['role'] == '감독':
            director_cnt += 1
    return director_cnt
            
```

```python
def check_lead_role_exist(cast):
    if not cast:
        return 0
    cast = eval((cast))
    lead_role_cnt = 0
    for person_dict in cast:
        if person_dict['role'] == '주연':
            lead_role_cnt += 1
    return lead_role_cnt
            
```

```python
def extend_director_string(cast):
    if not cast:
        return ''
    cast = eval((cast))
    director_lst = []
    for person_dict in cast:
        if person_dict['role'] == '감독':
            director_lst.append(person_dict['name'])
    return ', '.join(director_lst)
            
```

```python
def extend_lead_role_string(cast):
    if not cast:
        return ''
    cast = eval((cast))
    lead_role_lst = []
    for person_dict in cast:
        if person_dict['role'] == '주연':
            lead_role_lst.append(person_dict['name'])
    return ', '.join(lead_role_lst)
            
```

```python
def extend_supporting_role_string(cast):
    if not cast:
        return ''
    cast = eval((cast))
    supporting_role_lst = []
    for person_dict in cast:
        if person_dict['role'] == '출연':
            supporting_role_lst.append(person_dict['name'])
    return ', '.join(supporting_role_lst)
```
