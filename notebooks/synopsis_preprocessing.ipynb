# 테이블 생성 movidId, Synopsis

```python
with mysql.get_connection() as connection:
    cursor = connection.cursor()
    create_daum_movies_synopsis_prep_table_query = '''
        CREATE TABLE IF NOT EXISTS daum_movies_synopsis_prep (
            movieId INT PRIMARY KEY,
            synopsis_prep TEXT,
            FOREIGN KEY (movieId) REFERENCES daum_movies(movieId)
        )
    '''
    cursor.execute(create_daum_movies_synopsis_prep_table_query)
    
    # 변경 사항 커밋
    connection.commit()
    
    print("CREATE TABLE completed.")
```

```python
import re
bracket_regex = re.compile('\[')

def preprocess_synopsis(string):
    if not string:
        return ''
    search = bracket_regex.search(string)
    
    if not search:
        xa0_replaced = re.sub(r'\xa0+', ' ', synopsis)
    else:
        xa0_replaced = re.sub(r'\xa0+', ' ', synopsis[:search.start()])
        
    n_replaced = re.sub(r'\n+', ' ', xa0_replaced)
    ws_replaced = re.sub(r'\s+', ' ', n_replaced)

    return ws_replaced.strip()
```

```python
preps = []
for synopsis in df['synopsis']:
    preps.append(preprocess_synopsis(synopsis))
```

```python
df['preps'] = preps
```
